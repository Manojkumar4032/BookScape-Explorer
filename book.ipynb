```python
query = "history"  # Define the search query

# Initialize new_all_book_data as an empty list
book_scrape = []

# Process the data from data_list
for data in data_list:
    items = data.get('items', [])  # Get the list of items from the response

    # Check if 'items' exist and are not empty
    if not items:
        print("No items found in this response.")
        continue  # Skip to next iteration if no items are found

    for item in items:  # Loop through each item
        volume_info = item.get('volumeInfo', {})  # Get volumeInfo (book details)
        sale_info = item.get('saleInfo', {})  # Get saleInfo (price and availability)

        # Check if volume_info is empty
        if not volume_info:
            print(f"Volume info missing for item with ID: {item.get('id', 'N/A')}")
            continue  # Skip this item if no volumeInfo is available
        
        # Extract and store the book details
        book_info = {
            'book_id': item.get('id', ''),  # Book ID
            'search_key': query,  # The query used for the search
            'book_title': volume_info.get('title', ''),  # Book title
            'book_subtitle': volume_info.get('subtitle', ''),  # Book subtitle
            'book_authors': ", ".join(volume_info.get('authors', [])),  # Authors (comma-separated)
            'book_publisher':volume_info.get('publisher', ''),     #Book Publisher
            'book_description': volume_info.get('description', ''),  # Book description
            'industryIdentifiers': ", ".join([identifier.get('identifier', '') for identifier in volume_info.get('industryIdentifiers', [])]),  # Industry identifiers
            'text_readingModes': volume_info.get('readingModes', {}).get('text', False),  # Text reading mode availability
            'image_readingModes': volume_info.get('readingModes', {}).get('image', False),  # Image reading mode availability
            'pageCount': volume_info.get('pageCount', ''),  # Page count
            'categories': ", ".join(volume_info.get('categories', [])),  # Categories (comma-separated)
            'language': volume_info.get('language', ''),  # Language
            'imageLinks': volume_info.get('imageLinks', {}).get('thumbnail', ''),  # Thumbnail image link
            'ratingsCount': volume_info.get('ratingsCount', ''),  # Ratings count
            'averageRating': volume_info.get('averageRating', ''),  # Average rating
            'country': sale_info.get('country', ''),  # Sale info country
            'saleability': sale_info.get('saleability', ''),  # Saleability (e.g., if it's for sale)
            'isEbook': sale_info.get('isEbook', False),  # Is it available as an ebook?
            'amount_listPrice': sale_info.get('listPrice', {}).get('amount', ''),  # List price
            'currencyCode_listPrice': sale_info.get('listPrice', {}).get('currencyCode', ''),  # List price currency code
            'amount_retailPrice': sale_info.get('retailPrice', {}).get('amount', ''),  # Retail price
            'currencyCode_retailPrice': sale_info.get('retailPrice', {}).get('currencyCode', ''),  # Retail price currency code
            'buyLink': sale_info.get('buyLink', ''),  # URL to buy the book
            'year': volume_info.get('publishedDate', '').split('-')[0] if volume_info.get('publishedDate', '') else ''  # Year of publication (extracted from date)
            
        }

        # Append the book info to new_all_book_data
        book_scrape.append(book_info)

# Print the result
print(f"Total books fetched: {len(book_scrape)}")
```

```python
import numpy as np
# Assuming `new_all_book_data` is already populated with book information
# Convert the list of dictionaries to a Pandas DataFrame
df = pd.DataFrame(book_scrape)
#df.replace("",np.nan,inplace=True)
# Save the DataFrame to a CSV file
file_path = 'book_scrape.csv'  # Path where you want to save the file
df.to_csv(file_path, index=False)
print(f"CSV file saved at: {file_path}")
```

```python
import numpy as np
import pandas as pd

# Replace empty strings with NaN (use np.nan correctly)
df.replace("", np.nan, inplace=True)

# Debugging: Print column names and check for any leading/trailing spaces
#print(df.columns)

# Replace missing values with default values
df["book_subtitle"] = df["book_subtitle"].fillna("unknown")  # Fill missing book subtitles with "unknown"
df["book_authors"] = df["book_authors"].fillna("unknown")    # Fill missing authors with "unknown"
df["book_publisher"] = df["book_publisher"].fillna("unknown")  # Fill missing publishers with "unknown"
df["book_description"] = df["book_description"].fillna("unknown") # Fill missing descriptions with "unknown"
df["industryIdentifiers"] = df["industryIdentifiers"].fillna("0")  # Fill missing industry identifiers with "0"
df["pageCount"] = pd.to_numeric(df["pageCount"], errors='coerce').fillna(0).astype(int)  # Convert pageCount to integer
df["categories"] = df["categories"].fillna("unknown")  # Fill missing categories with "unknown"
df["imageLinks"] = df["imageLinks"].fillna("unknown")  # Fill missing image links with "unknown"
df["ratingsCount"] = pd.to_numeric(df["ratingsCount"], errors='coerce').fillna(0).astype(int)  # Convert ratingsCount to integer
df["averageRating"] = pd.to_numeric(df["averageRating"], errors='coerce').fillna(0).astype(float)  # Convert averageRating to float
df["amount_listPrice"] = pd.to_numeric(df["amount_listPrice"], errors='coerce').fillna(0).astype(float)  # Convert amount_listPrice to float
df["currencyCode_listPrice"] = df["currencyCode_listPrice"].fillna("unknown")  # Fill missing currency codes with "unknown"
df["amount_retailPrice"] = pd.to_numeric(df["amount_retailPrice"], errors='coerce').fillna(0).astype(float)  # Convert amount_retailPrice to float
df["currencyCode_retailPrice"] = df["currencyCode_retailPrice"].fillna("unknown")  # Fill missing retail currency codes with "unknown"
df["buyLink"] = df["buyLink"].fillna("unknown")  # Fill missing buy links with "unknown"
df["year"] = df["year"].fillna("unknown")  # Fill missing years with "unknown"

# Verify if missing values were filled
print(df.isnull().sum())
```
